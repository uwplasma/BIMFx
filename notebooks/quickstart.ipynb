# BIMFx Quickstart

This notebook demonstrates a minimal solve + validation workflow.

```python
import numpy as np
from bimfx import solve_mfs
from bimfx.io import load_boundary
from bimfx.validation import relative_boundary_residual, summary_stats

data = load_boundary("../inputs/knot_tube.csv", normals_path="../inputs/knot_tube_normals.csv")
P, N = data.points, data.normals

field = solve_mfs(P, N, toroidal_flux=1.0)
res = relative_boundary_residual(field.B, P, N)
summary_stats(res)
```
